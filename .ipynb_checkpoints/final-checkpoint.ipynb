```python
data = {
    'carbohydrates_(g)': [26.0, 27.0],
    'cholesterol_(mg)': [0, 5],
    'energy_(kcal)': [150, 375],
    'fat_(g)': [0.5, 15.0],
    'fiber_(g)': [1.0, 2.2],
    'minerals_(mg)': [10, 300],  # mostly sodium, iron, calcium in noodles
    'proteins_(g)': [3.0, 7.0],
    'salt/sodium_(mg)': [5, 900],
    'saturated-fat_(g)': [0.1, 6.0],
    'vitamins_(mg)': [0.1, 1.5],  # includes some added B-vitamins in Maggi
    'nova_group': [3, 4]  # Vermicelli is processed; instant noodles are ultra-processed
}
df2 = pd.DataFrame(data)
df = pd.concat([df, df2], ignore_index=True)
df.tail(5)
```

```python
# Let's say you want 'nova_group' to be the last column
col_to_end = 'nova_group'

# Get all other columns and sort them
ordered_cols = sorted([col for col in df.columns if col != col_to_end])

# Add the specified column to the end
new_column_order = ordered_cols + [col_to_end]

# Reorder the DataFrame
df = df[new_column_order]
```

```python
df.sample(2)
```

```python
# Cell 3: Separate features and target
X = df.drop(columns=["nova_group"])
y = df["nova_group"]

print("Total Features:", X.shape[1])
```

```python
df.head(3)
```

```python
# Cell 4: Stratified sampling of 50,000 rows
X_sample, _, y_sample, _ = train_test_split(
    X,
    y,
    train_size=50000,
    stratify=y,
    random_state=42
)

df_sampled = pd.concat([X_sample, y_sample], axis=1)
df_sampled.to_csv("stratified_sampled_nova_data.csv", index=False)

print("Sampled Data Shape:", df_sampled.shape)
print("Class Distribution:\n", y_sample.value_counts(normalize=True))
```

```python
# Cell 5: Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X_sample,
    y_sample,
    test_size=0.2,
    random_state=42,
    stratify=y_sample
)
```

```python
df
```

```python
# Cell 6: Train smaller Random Forest model
model = RandomForestClassifier(
    n_estimators=50,
    max_depth=10,
    random_state=42
)
model.fit(X_train, y_train)
```

```python

from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt


# Get learning curve data
train_sizes, train_scores, val_scores = learning_curve(
    estimator=model,
    X=X,
    y=y,
    cv=5,                  # 5-fold cross-validation
    train_sizes=np.linspace(0.1, 1.0, 10),  # Vary training set size from 10% to 100%
    scoring='accuracy',    # or 'f1', 'roc_auc', etc. for classification
    n_jobs=-1              # Use all available CPU cores
)

# Calculate mean and std deviation for plotting
train_scores_mean = train_scores.mean(axis=1)
val_scores_mean = val_scores.mean(axis=1)

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(train_sizes, train_scores_mean, 'o-', label="Training score")
plt.plot(train_sizes, val_scores_mean, 'o-', label="Cross-validation score")
plt.xlabel("Training Set Size")
plt.ylabel("Accuracy")
plt.title("Learning Curve")
plt.legend(loc="best")
plt.grid()
plt.tight_layout()
plt.savefig("learning_curve.jpg", format='jpg', dpi=300)
plt.show()
```

```python
from scipy.stats import pearsonr
import pandas as pd

# List of features you want to test (excluding target and Nutri/NOVA scores)
features = ['carbohydrates', 'cholesterol', 'energy', 'fat', 'fiber',
            'proteins', 'salt', 'saturated-fat', 'minerals', 'vitamins']

results = []

for feature in features:
    r, p = pearsonr(df[feature], df['nova_group'])
    result = "Significant (Reject H₀)" if p < 0.05 else "Not Significant (Fail to Reject H₀)"
    results.append((feature, round(r, 4), "{:.4e}".format(p), result))

# Create DataFrame to display results nicely
results_df = pd.DataFrame(results, columns=['Feature', 'Correlation (r)', 'p-value', 'Result'])

print(results_df)

```

```python
# Cell 7: Evaluate the model
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```
